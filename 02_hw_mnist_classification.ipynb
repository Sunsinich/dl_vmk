```python
device = torch.device('cuda' if torch.cuda.is_available() else 'cpu')
model = model.to(device)
```

```python
criterion = nn.CrossEntropyLoss()
optimizer = torch.optim.Adam(model.parameters(), lr = 0.001)# your code here
```

```python
def training(model, train_loader, test_loader, criterion, optimizer, num_epoch = 20):
  for epoch in range(num_epoch):
    model.train()

    for ind, (data, target) in enumerate(train_loader):
      data, target = data.to(device), target.to(device)

      optimizer.zero_grad()
      pred = model(data)

      loss = criterion(pred, target)
      loss.backward()

      optimizer.step()
    print(epoch)
```

```python
training(model, train_data_loader, test_data_loader, criterion, optimizer)
```

Оценим качество классификации:

```python
predicted_labels = []
real_labels = []
model.eval()
with torch.no_grad():
    for batch in train_data_loader:
        data = batch[0].to(device)
        target = batch[1].to(device)
        y_predicted = model(data.reshape(-1, 784))
        predicted_labels.append(y_predicted.argmax(dim=1))
        real_labels.append(target)

predicted_labels = torch.cat(predicted_labels)
real_labels = torch.cat(real_labels)
train_acc = (predicted_labels == real_labels).type(torch.FloatTensor).mean()
```
